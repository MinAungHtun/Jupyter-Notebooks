```python
data = pd.read_csv(r"C:\Users\acer\Downloads\potential-enigma-master\potential-enigma-master\headbrain.csv")
print(data.shape)
data.head()
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
n=len(X)
numer=0
denom=0
for i in range(n):
    numer +=(X[i]-mean_x)*(Y[i]-mean_y)
    denom +=(X[i]-mean_x)**2
b1=numer/denom
b0=mean_y-(b1*mean_x)
print(b1,b0)
```

```python
max_x=np.max(X)+100
min_x=np.min(X)-100
x=np.linspace(min_x,max_x,1000)
y=b0+b1*x
plt.plot(x,y,color='#58b970', label='Regression Line')
plt.scatter(X,Y,c='#ef5423', label='Scatter Plot')
plt.xlabel('HeadSize in cm3')
plt.ylabel('Brainweight in grams')
plt.legend()
plt.show()
```

```python
ss_t=0
ss_r=0
for i in range (n):
    y_pred=b0+b1*X[i]
    ss_t +=(Y[i]-mean_y)**2
    ss_r +=(Y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=X.reshape((n,1))
reg=LinearRegression()
reg=reg.fit(X,Y)
Y_pred=reg.predict(X)
r2_score=reg.score(X,Y)
print (r2_score)
```
