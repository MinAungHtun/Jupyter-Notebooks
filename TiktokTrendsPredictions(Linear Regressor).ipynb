```python
X_trends = tiktok_trends[['danceability', 'energy', 'loudness',
       'speechiness','acousticness', 'instrumentalness', 'liveness',
     'tempo', 'duration_mins']] # Valence was dropped due to giving better clf scores for the linear regressor when dropped
X_trends = pd.get_dummies(X_trends)
Y_trends = tiktok_trends[['popularity']]
```

```python
X_trends.to_csv('training_data_without_valence.csv')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_trends, Y_trends, test_size=0.30, random_state=15)
```

```python
X_train.value_counts()
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression().fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)
Y_pred
```

```python
clf.score(X_test, Y_test)
```
